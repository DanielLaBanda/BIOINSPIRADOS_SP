```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import time # Importar el m√≥dulo time

# =============================================================================
# 1. CONFIGURACI√ìN
# =============================================================================
ARCHIVO_EXCEL = 'matriz_distancias_tsp.xlsx'
TAMANOS_REDES = [100, 150, 200]
CANTIDAD_POR_TAMANO = 10
SEED = 42

random.seed(SEED)
np.random.seed(SEED)

# =============================================================================
# 2. PROCESAMIENTO DE DATOS (MISMA FUNCI√ìN QUE USAS PARA ACO)
# =============================================================================

def cargar_y_normalizar_matriz(filepath):
    print(f"üìÇ Cargando archivo: {filepath} ...")
    try:
        df = pd.read_excel(filepath, header=None, index_col=None)
        D = df.values.astype(float)
    except Exception as e:
        print(f"‚ùå Error al cargar Excel: {e}")
        return None

    print(f"   Dimensi√≥n original: {D.shape}")

    # 1. Diagonal = 0
    np.fill_diagonal(D, 0)

    # 2. Hacer sim√©trica
    D = (D + D.T) / 2

    # 3. Reemplazar NaNs por un valor grande
    max_val = np.nanmax(D) if not np.isnan(D).all() else 1000
    D = np.nan_to_num(D, nan=max_val)

    # 4. Outliers (percentil 99)
    perc99 = np.percentile(D, 99)
    D[D > perc99] = perc99

    # 5. Escalar a [0,100]
    if D.max() != 0:
        D = (D / D.max()) * 100

    # 6. Evitar ceros fuera de diagonal
    mask_zero = (D < 0.01) & (np.eye(D.shape[0]) == 0)
    D[mask_zero] = 0.01

    print("‚úÖ Matriz maestra normalizada lista.")
    return D


def generar_lotes_redes(matriz_maestra, sizes, n_per_size):
    """Genera submatrices listas para TSP."""
    lote = []
    total_nodos = matriz_maestra.shape[0]

    for s in sizes:
        if s > total_nodos:
            print(f"‚ö†Ô∏è No se puede generar red de {s} nodos.")
            continue

        for i in range(n_per_size):
            indices_reales = sorted(random.sample(range(total_nodos), s))
            submatriz = matriz_maestra[np.ix_(indices_reales, indices_reales)]

            lote.append({
                "id_interno": f"{s}_nodos_v{i+1}",
                "n_nodos": s,
                "indices_reales": indices_reales,
                "matriz_dist": submatriz
            })
    return lote


# =============================================================================
# 3. FUNCIONES DEL ALGORITMO GEN√âTICO
# =============================================================================

def tour_length(tour, dist):
    total = 0.0
    n = len(tour)
    for i in range(n - 1):
        total += dist[tour[i], tour[i + 1]]
    total += dist[tour[-1], tour[0]]
    return total


def init_population(pop_size, n):
    pop = []
    base = list(range(n))
    for _ in range(pop_size):
        indiv = base[:]
        random.shuffle(indiv)
        pop.append(indiv)
    return pop


def tournament_selection(pop, fitness, k=3):
    indices = random.sample(range(len(pop)), k)
    best = min(indices, key=lambda i: fitness[i])
    return pop[best][:]


def order_crossover(p1, p2):
    n = len(p1)
    a, b = sorted(random.sample(range(n), 2))
    child = [None] * n
    child[a:b+1] = p1[a:b+1]

    pos = (b + 1) % n
    for gene in p2:
        if gene not in child:
            child[pos] = gene
            pos = (pos + 1) % n
    return child


def swap_mutation(tour, mutation_rate):
    if random.random() < mutation_rate:
        i, j = random.sample(range(len(tour)), 2)
        tour[i], tour[j] = tour[j], tour[i]
    return tour


def evolve_one_network(dist,
                       pop_size=200,
                       generations=400,
                       crossover_rate=0.9,
                       mutation_rate=0.2,
                       tournament_k=3,
                       verbose=True):

    n = dist.shape[0]
    pop = init_population(pop_size, n)

    fitness = [tour_length(ind, dist) for ind in pop]
    best_idx = min(range(pop_size), key=lambda i: fitness[i])
    best_tour = pop[best_idx][:]
    best_fit = fitness[best_idx]

    history = [best_fit]

    for gen in range(1, generations + 1):
        new_pop = []

        while len(new_pop) < pop_size:
            p1 = tournament_selection(pop, fitness, tournament_k)
            p2 = tournament_selection(pop, fitness, tournament_k)

            child = order_crossover(p1, p2) if random.random() < crossover_rate else p1[:]
            child = swap_mutation(child, mutation_rate)

            new_pop.append(child)

        pop = new_pop
        fitness = [tour_length(ind, dist) for ind in pop]

        gen_best_idx = min(range(pop_size), key=lambda i: fitness[i])
        gen_best_fit = fitness[gen_best_idx]

        if gen_best_fit < best_fit:
            best_fit = gen_best_fit
            best_tour = pop[gen_best_idx][:]

        history.append(best_fit)

        if verbose and gen % 50 == 0:
            print(f"   Gen {gen} | Mejor costo: {best_fit:.2f}")

    return best_tour, best_fit, history


# =============================================================================
# 4. GRAFICAR
# =============================================================================

def graficar_convergencia(history, title):
    plt.figure()
    plt.plot(history)
    plt.title(title)
    plt.xlabel("Generaci√≥n")
    plt.ylabel("Mejor costo")
    plt.grid(True)
    plt.show()


# =============================================================================
# 5. EJECUCI√ìN PRINCIPAL DEL GA
# =============================================================================

matriz_maestra = cargar_y_normalizar_matriz(ARCHIVO_EXCEL)

redes = generar_lotes_redes(
    matriz_maestra,
    TAMANOS_REDES,
    CANTIDAD_POR_TAMANO
)

resultados = []

for r in redes:
    print("\n======================================")
    print(f"üîç Evaluando red: {r['id_interno']} ({r['n_nodos']} nodos)")
    print("======================================")

    dist = r["matriz_dist"]

    start_time = time.time() # Iniciar cron√≥metro
    best_tour, best_fit, history = evolve_one_network(
        dist,
        pop_size=200,
        generations=400,
        crossover_rate=0.9,
        mutation_rate=0.2,
        tournament_k=3,
        verbose=False
    )
    end_time = time.time() # Detener cron√≥metro
    execution_time = end_time - start_time # Calcular tiempo de ejecuci√≥n

    print(f"üèÅ Mejor costo GA: {best_fit:.2f}")
    print(f"‚è±Ô∏è Tiempo de ejecuci√≥n: {execution_time:.2f} segundos")

    resultados.append({
        "id": r["id_interno"],
        "size": r["n_nodos"],
        "best_fit": best_fit,
        "best_tour": best_tour,
        "history": history,
        "execution_time": execution_time # Guardar el tiempo de ejecuci√≥n
    })

    graficar_convergencia(history, f"Convergencia GA - {r['id_interno']}")

# =============================================================================
# 6. AN√ÅLISIS DE RESULTADOS POR TAMA√ëO DE RED
# =============================================================================

print("\n======================================")
print("üìä Resumen de Resultados por Tama√±o de Red")
print("======================================")

resultados_df = pd.DataFrame(resultados)
summary_df = resultados_df.groupby('size').agg(
    avg_best_fit=('best_fit', 'mean'),
    avg_execution_time=('execution_time', 'mean')
).reset_index()

for index, row in summary_df.iterrows():
    print(f"\n--- Redes de {int(row['size'])} Nodos ---")
    print(f"  Promedio del Mejor Costo (GA): {row['avg_best_fit']:.2f}")
    print(f"  Promedio del Tiempo de Ejecuci√≥n: {row['avg_execution_time']:.2f} segundos")
```
